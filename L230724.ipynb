```python
import glob
def FileCollection(s_paths, file_types):
    dummy_arg = s_paths + '**/*.'+file_types
    print(dummy_arg)
    file_lists = glob.glob(dummy_arg, recursive=True)
    return file_lists
```

```python
def GetFileSentence(file_list, search_sentences, s_path):
    file_dict = {}
    for filename in file_list:
        #print(filename)
        sentence_list = []
        with open(filename, encoding='utf-8') as f:
            lines = f.readlines()
        for item in lines:
            if search_sentence in item:
                sentence_list.append(item)
        if sentence_list :
            file_dict[filename[len(s_path):]] = sentence_list
    return(file_dict)
```
