# Query 함수(판다스)

```python
data = {'age':[10, 10, 21, 22], 'weight':[20, 30, 60, 70]}
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.query('age == 10 ')
```

```python
expr_str = 'age == 10'
```

```python
df.query(expr_str)
```

```python
expr_str = 'age != 10'
df.query(expr_str)
```

```python
expr_str = 'age > 20'
df.query(expr_str)
```

```python
expr_str = 'age == [21, 22]'
df.query(expr_str)
```

```python
expr_str = 'age not in [21, 22]'
df.query(expr_str)
```

```python
expr_str = '(age == 10) and (weight >= 30)'
df.query(expr_str)
```

'(age == @num_age) and (weight >= @num_weight)'

```python
num_age = 10
num_weight = 30
expr_str = f'(age == {num_age} and weight >= {num_weight})'
df.query(expr_str)
```

```python
def max_user(x, y):
    return max(x, y)
```

```python
max_user(1, 22)
```

```python
expr_str = 'age >= @max_user(1, 22)'
df.query(expr_str)
```

```python
max(1, 22)
```

```python
expr_str = 'age >= max(1, 22)'
df.query(expr_str)
```

```python
df
```

```python
expr_str = 'index > 1'
df.query(expr_str)
```

```python
data = {'name':['White tiger', 'Tiger black', 'Red tiger'], 'age':[5, 7, 9]}
df = pd.DataFrame(data)
```

```python
df
```

```python
expr_str = '''name.str.contains('tiger')'''
df.query(expr_str, engine='python')
```

```python
expr_str = '''name.str.contains('tiger', case=False)'''
df.query(expr_str, engine='python')
```

```python
expr_str = '''name.str.startswith('Tiger')'''
df.query(expr_str, engine='python')
```

```python
expr_str = '''name.str.endswith('tiger')'''
df.query(expr_str, engine='python')
```
