```python
def put_up(df): #df <- df_float
    column_list = list(df.columns)
    df_list = []
    for i in column_list[4:]:
        mask = df[i].notnull()
        col_list = column_list[:4]+[i] # ['식품코드', 'DB군', '식품명', '지역/제조사', '수분...']
        #print(col_list)
        df_ = df[mask].loc[:, col_list]
        df_ = df_.rename(columns={i:'value'})
        df_['nutrients']=i
        df_list.append(df_)
    nutrients = pd.concat(df_list)
    nutrients.reset_index(drop=True, inplace=True)
    return (nutrients)
```

```python
from matplotlib import font_manager
def get_data(fn, nu):
    nu_list = nu[0:-2] + list(range(nu[-2], nu[-1]+1))
    df = pd.read_excel(fn, engine='openpyxl', usecols=nu_list)
    df = df.replace('-', np.nan)
    return(df)


def df_plot(df): #df <- df_mean
    font_path='./drive/MyDrive/NanumGothic.ttf'
    fontprop = font_manager.FontProperties(fname=font_path, size=20)
    plt.figure(figsize=(6, 6))
    df = df.sort_values(ascending=False)
    plt.bar(df.index[:10], df[:10])
    plt.xticks(fontproperties=fontprop, rotation='vertical')
    plt.tight_layout()
    plt.savefig('marine.jpg')

def means(df):
    dict_dtype = {}
    for i in df.columns[4:]:
        dict_dtype[i]='float'
    df2=df.astype(dict_dtype)
    df_mean=df2.iloc[:, 4:].mean()
    df_plot(df_mean)
    return(df2)

def put_up(df): #df <- df_float
    df_list = []
    for i in df.columns[4:]:
        df_ = pd.concat([df.iloc[:, :4], df[i]], axis=1)
        df_ = df_.dropna()
        df_ = df_.rename(columns={i:'value'})
        df_['nutrients']=i
        df_list.append(df_)
    nutrients = pd.concat(df_list)
    nutrients.reset_index(drop=True, inplace=True)
    return (nutrients)
```
