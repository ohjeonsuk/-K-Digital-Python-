완전 탐색 = BFA

x : 닭의 수

y : 돼지의 수


x : 0 - 20 (-100, 1000)

y : 0 - 20


x = 0, y = 20, 다리의 개수가 맞는지 검사

x = 1, y = 19, 다리의 개수가 맞는지 검사

....



x + y + z = 20

2x + 4y + 8z = 56

```python
def brute_force_algorithm(total_head, total_leg):
    '''
    i : 다리가 2개인 동물의 수
    j : 다리가 4개인 동물의 수
    k : 다리가 8개인 곤충의 수 (거미)
    '''
    for i in range(total_head + 1):
        j = total_head - i
        condition = 2*i + 4*j
        if condition == total_leg:
            return(i, j)
    return(None, None)
```

```python
brute_force_algorithm(56, 20)
```

```python
def barnyard():
    total_head = int(input('총 마리수: '))
    total_leg = int(input('총 다리수: '))

    chicken, pig = brute_force_algorithm(total_head, total_leg) # i == 12 ==> chicken, j == 8 ==> pig
    if any((chicken, pig)):
        print(f'닭 수: {chicken}, 돼지 수: {pig}')
    else:
        print('There is no solution.')
```

```python
def brute_force_algorithm(total_head, total_leg):
    for x in range(total_head +1):
        for y in range(total_head - x + 1):
            z = total_head - x - y
            if total_leg == 2*x + 4*y +8*z:
                return x, y, z
    return(None, None, None)
```

```python
def brute_force_algorithm(total_head, total_leg):
    result = []
    for x in range(total_head +1):
        for y in range(total_head - x + 1):
            z = total_head - x - y
            if total_leg == 2*x + 4*y +8*z:
                result.append((x, y, z))
    return(result)
```

```python
def barnyard():
    total_head = int(input("총 머릿수를 입력 해주 주세요."))
    total_leg = int(input("총 다리 수를 입력 해주 주세요."))

    result = brute_force_algorithm(total_head, total_leg)
    # result = [(1, 2, 3), (3, 4, 5), ... ]
    print(result)

    if result :
        for i in result:
            print(f"닭 수 : {i[0]},  돼지 수 : {i[1]},  거미 수 : {i[2]}")
    else:
        print("There is no solution")
```
